```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/kaggle/input/gender-by-name/name_gender_dataset.csv')
df = df.dropna(axis=1)
df.shape
```

```python
df["Gender"] = df["Gender"].map({"M":1,"F":0})
```

```python
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df["Name"] = encoder.fit_transform(df["Name"])
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

x=df.drop('Gender',axis=1)
y=df['Gender']
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.25, random_state=42)
```

```python
x_train.head()
```

# **LOGISTIC REGRESSION**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn import metrics

LGR_Classifier = LogisticRegression(n_jobs=-1, random_state=0)
LGR_Classifier.fit(x_train, y_train)
predictions = LGR_Classifier.predict(x_test)
accuracy_lr = accuracy_score(y_test, predictions)

print("Model Accuracy : {}".format(accuracy_lr))

confusion_matrix = metrics.confusion_matrix(y_test, predictions)
print("Confusion matrix:" "\n", confusion_matrix)
```

# **SVM**

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn import metrics
from sklearn.metrics import classification_report


svc_clf = LinearSVC()
svc_clf.fit(x_train, y_train)
predictions = svc_clf.predict(x_test)
accuracy_svm = accuracy_score(y_test,predictions)
print(accuracy_svm)
confusion_matrix = metrics.confusion_matrix(y_test, predictions)
print("Confusion matrix:" "\n", confusion_matrix)
```

# **DECISION TREE**

```python
# DECISION TREE CLASSIFIER
from sklearn.tree import DecisionTreeClassifier

deccision_classifier = DecisionTreeClassifier()
deccision_classifier.fit(x_train,y_train)
```

```python
# accuracy
# confusion matrix
predictions = deccision_classifier.predict(x_test)
accuracy_dt = metrics.accuracy_score(y_test, predictions)
confusion_matrix = metrics.confusion_matrix(y_test,predictions)

print ("Decision Classifier Model Accuracy:" "\n", accuracy_dt)
print("Decision Classifier Model Confusion matrix:" "\n", confusion_matrix)
```
