# map
gives the array  common between two arrays (vlookup)

```python
a=pd.Series(["goa","bnglr","andhra","kerala"],index=["tulu","kannada","telugu","malayalam"])
a
```

```python
k=pd.Series(["tulu","kannada","telugu","malayalam"],index=["panaji","kartanaka","vizag","tpuram"])
k
```

```python
k.map(a)
```
